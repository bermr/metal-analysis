```python
#metal bands analysis

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

world = pd.read_csv('datasets/world_population_1960_2015.csv', encoding='latin-1')
bands = pd.read_csv('datasets/metal_bands_2017.csv', encoding='latin-1')
```

```python
# ammount of band per country

bands_country = bands['origin'].value_counts()

plt.figure(figsize = (12,6))
plt.title('Counts of bands per country')
sns.barplot(x=bands_country[:10].keys(), y=bands_country[:10].values, palette=sns.color_palette("tab20c"))
plt.show()
```

```python
# ammount of bands by style

bands_styles = bands['style'].value_counts()

plt.figure(figsize = (12,6))
plt.title('Counts of bands per style')
sns.barplot(x=bands_styles[:10].keys(), y=bands_styles[:10].values, palette=sns.color_palette("tab20c"))
plt.show()
```

```python
# ammount of bands by year

bands_formed = bands['formed'].value_counts()

bands_formed.pop('-')
plt.figure(figsize = (12,6))
plt.title('Bands created per year')
sns.lineplot(x=bands_formed.keys(), y=bands_formed.values, color='black')
plt.show()
```

```python
# ammount of splitted bands by year

bands_splitted = bands['split'].value_counts()

bands_splitted.pop('-')
plt.figure(figsize = (12,6))
plt.title('Bands splitted per year')
sns.lineplot(x=bands_splitted.keys(), y=bands_splitted.values, color='black')
plt.show()
```

```python
# bands / population ratio

world_pop = world['2015']
bands_country = bands['origin'].value_counts()
band_count = [0 for i in range(len(world))]

for i, country in world.iterrows():
    if country['Country Name'] in bands_country:
        band_count[i] = bands_country[country['Country Name']]

        
world['mpc'] = (band_count / world_pop)
world = world.sort_values(by = 'mpc', ascending=False)

plt.figure(figsize = (12,6))
plt.title('Bands per capita ratio')
sns.barplot(x=world[:10]['Country Name'], y=world[:10]['mpc'], palette=sns.color_palette("tab20c"))
plt.show()
```
