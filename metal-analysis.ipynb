```python
#metal bands analysis

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

world = pd.read_csv('datasets/world_population_1960_2015.csv', encoding='latin-1')
bands = pd.read_csv('datasets/metal_bands_2017.csv', encoding='latin-1')
```

```python
# ammount of band per country

bands_country = bands['origin'].value_counts()

plt.figure(figsize = (12,6))
plt.title('Counts of bands per country')
sns.barplot(x=bands_country[:10].keys(), y=bands_country[:10].values, palette=sns.color_palette("tab20c"))
plt.show()
```

```python
# ammount of bands by style

bands_styles = bands['style'].value_counts()

plt.figure(figsize = (12,6))
plt.title('Counts of bands per style')
sns.barplot(x=bands_styles[:10].keys(), y=bands_styles[:10].values, palette=sns.color_palette("tab20c"))
plt.show()
```

```python
# ammount of bands by year

bands_years = bands['formed'].value_counts()

plt.figure(figsize = (12,6))
plt.title('Counts of bands created per year')
sns.barplot(x=bands_years[:10].keys(), y=bands_years[:10].values, palette=sns.color_palette("Dark2"))
plt.show()
```
